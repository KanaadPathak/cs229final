```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
```

```python
# Load CSV using Pandas
df = pd.read_csv('data/train.csv')
print(df.shape)
```

```python
textures = df.filter(['texture1', 'texture2', 'texture3'])
scatter_matrix(textures)
plt.show()
```

```python
margins = df.filter(['margin1', 'margin2', 'margin3', 'margin4'])
scatter_matrix(margins)
plt.show()
```

```python
selected = ["shape%i" % i for i in range(1, 10)]
scatter_matrix(df.filter(selected))
plt.show()
```
